```python
import json

with open("./doughnut1000tsne/doughnut1000.json") as json_file:
    donut_data = json.load(json_file)

donuts = dict()
for item in donut_data:
    donuts[item['path']] = item['point']
```

```python
import math
def distance(coord1, coord2):
    return math.sqrt(sum([(i - j)**2 for i, j in zip(coord1, coord2)]))
```

```python
def closest(space, coord, n=5):
    closest = []
    for key in sorted(space.keys(),
                      key=lambda x: distance(coord, space[x]))[1:n+1]:
        closest.append(key)
    return closest

import random
# key_donut = random.choice(list(donuts.keys()))
# similar_donuts = closest(donuts, donuts[key_donut])
key_donut = donuts['/Users/hksun93/thesis/doughnut1000/598360.png']
similar_donuts = closest(donuts, key_donut)
similar_donuts
```

```python
from IPython.display import Image, display

display_donuts = similar_donuts

for donut_image in display_donuts:
    display(Image(filename = donut_image))
```
